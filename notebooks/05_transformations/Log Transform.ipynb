```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

path = '/content/drive/MyDrive/Image Processing Lab Files/MST Lab File (.ipynb)/Tiger.png'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

img = img.astype(np.float64)
img = img / 255.0

c = 255 / np.log(1 + np.max(img) + 1e-5)

log_transform = c * np.log(1 + img)

log_transform = np.nan_to_num(log_transform)
log_transform = np.array(np.clip(log_transform, 0, 255), dtype=np.uint8)
img = img * 255.0

cv2_imshow(img.astype(np.uint8))
cv2_imshow(log_transform)
```
