```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

path = '/content/drive/MyDrive/Image Processing Lab Files/MST Lab File (.ipynb)/Tiger.png'
tiger = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

norm = tiger / 255.0
g = float(input("Enter gamma: "))
gamma_img = np.power(norm, g)

gamma_img = (gamma_img * 255.0).astype(np.uint8)

cv2_imshow(tiger)
cv2_imshow(gamma_img)
```
