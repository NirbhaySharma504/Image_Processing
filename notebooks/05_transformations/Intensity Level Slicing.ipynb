```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

path = '/content/drive/MyDrive/Image Processing Lab Files/MST Lab File (.ipynb)/Tiger.png'
tiger = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

low_in = int(input("Enter r1 (lower input): "))
high_in = int(input("Enter r2 (upper input): "))
low_out = int(input("Enter s1 (lower output): "))
high_out = int(input("Enter s2 (upper output): "))

result = tiger.copy().astype(np.float32)

rng = (tiger >= low_in) & (tiger <= high_in)
result[rng] = low_out + (tiger[rng] - low_in) * ((high_out - low_out) / (high_in - low_in))

result = np.clip(result, 0, 255).astype(np.uint8)

cv2_imshow(tiger)
cv2_imshow(result)
```
