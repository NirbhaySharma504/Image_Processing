```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

path = '/content/drive/MyDrive/Image Processing Lab Files/EST Lab File (.ipynb)/Eagle.png'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
h, w = img.shape

lbp = np.zeros((h, w), dtype=np.uint8)

for i in range(1, h-1):
    for j in range(1, w-1):
        center = img[i, j]
        binary = []

        binary.append(1 if img[i-1, j-1] >= center else 0)
        binary.append(1 if img[i-1, j]   >= center else 0)
        binary.append(1 if img[i-1, j+1] >= center else 0)
        binary.append(1 if img[i, j+1]   >= center else 0)
        binary.append(1 if img[i+1, j+1] >= center else 0)
        binary.append(1 if img[i+1, j]   >= center else 0)
        binary.append(1 if img[i+1, j-1] >= center else 0)
        binary.append(1 if img[i, j-1]   >= center else 0)

        value = 0
        for k in range(8):
            value += binary[k] << (7 - k)

        lbp[i, j] = value

cv2_imshow(img)
cv2_imshow(lbp)
```
