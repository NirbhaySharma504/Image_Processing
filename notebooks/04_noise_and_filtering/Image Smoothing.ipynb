```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

path = '/content/drive/MyDrive/Image Processing Lab Files/MST Lab File (.ipynb)/Tiger.png'
tiger = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

avg_kernel = np.ones((7,7), dtype=np.float32) / 49
avg_out = cv2.filter2D(tiger, -1, avg_kernel)

gauss_out = cv2.GaussianBlur(tiger, (11,11), 3)

w_kernel = np.zeros((3,3), dtype=np.float32)
for i in range(3):
    for j in range(3):
        w_kernel[i, j] = max(abs(i-1), abs(j-1))
w_kernel = w_kernel / w_kernel.sum()
w_out = cv2.filter2D(tiger, -1, w_kernel)

cv2_imshow(avg_out)
cv2_imshow(w_out)
cv2_imshow(gauss_out)

```
