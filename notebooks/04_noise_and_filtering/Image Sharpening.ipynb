```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

path = '/content/drive/MyDrive/Image Processing Lab Files/EST Lab File (.ipynb)/Eagle.png'
eagle = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

lap = cv2.Laplacian(eagle, cv2.CV_64F, ksize=5)
lap = cv2.convertScaleAbs(lap)

sobel_x = cv2.Sobel(eagle, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(eagle, cv2.CV_64F, 0, 1, ksize=5)
sobel = cv2.convertScaleAbs(sobel_x) + cv2.convertScaleAbs(sobel_y)

canny = cv2.Canny(eagle, 50, 150)

blur = cv2.GaussianBlur(eagle, (9,9), 5)
unsharp = cv2.addWeighted(eagle, 1.8, blur, -0.8, 0)

cv2_imshow(lap)
cv2_imshow(sobel)
cv2_imshow(canny)
cv2_imshow(unsharp)

```
