```python
import cv2
import numpy as np
from scipy.signal import wiener
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

path = '/content/drive/MyDrive/Image Processing Lab Files/MST Lab File (.ipynb)/Tiger.png'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE) / 255.0

gauss_noise = img + np.random.normal(0, 0.25, img.shape)
gauss_noise = np.clip(gauss_noise, 0, 1)

speckle_noise = img + img * np.random.normal(0, 0.4, img.shape)
speckle_noise = np.clip(speckle_noise, 0, 1)

wiener_gauss = wiener(gauss_noise, (3, 3))
wiener_speckle = wiener(speckle_noise, (3, 3))

median_gauss = cv2.medianBlur((gauss_noise*255).astype(np.uint8), 3)
median_speckle = cv2.medianBlur((speckle_noise*255).astype(np.uint8), 3)

gaussian_gauss = cv2.GaussianBlur(gauss_noise, (3,3), 1)
gaussian_speckle = cv2.GaussianBlur(speckle_noise, (3,3), 1)

bilateral_gauss = cv2.bilateralFilter((gauss_noise*255).astype(np.uint8), 5, 75, 75)
bilateral_speckle = cv2.bilateralFilter((speckle_noise*255).astype(np.uint8), 5, 75, 75)

cv2_imshow((gauss_noise*255).astype(np.uint8))
cv2_imshow((speckle_noise*255).astype(np.uint8))
cv2_imshow((wiener_gauss*255).astype(np.uint8))
cv2_imshow((wiener_speckle*255).astype(np.uint8))
cv2_imshow(median_gauss)
cv2_imshow(median_speckle)
cv2_imshow((gaussian_gauss*255).astype(np.uint8))
cv2_imshow((gaussian_speckle*255).astype(np.uint8))
cv2_imshow(bilateral_gauss)
cv2_imshow(bilateral_speckle)
```
