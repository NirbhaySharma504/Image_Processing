```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.util import random_noise
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

path = '/content/drive/MyDrive/Image Processing Lab Files/EST Lab File (.ipynb)/Tiger.png'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE) / 255.0

a, b = 0, 1
uniform_noise = np.random.uniform(a, b, img.shape)

mu, sigma = 0, 0.1
gaussian_noise = np.random.normal(mu, sigma, img.shape)
gaussian_builtin = random_noise(img, mode='gaussian', mean=mu, var=sigma**2)

k = 3
lam = 2
erlang_noise = np.random.gamma(k, 1/lam, img.shape)

lam_exp = 2
exp_noise = np.random.exponential(1/lam_exp, img.shape)

sigma_r = 0.5
rayleigh_noise = sigma_r * np.sqrt(-2 * np.log(1 - np.random.rand(*img.shape)))

noise_data = [
    ("Uniform", uniform_noise),
    ("Gaussian", gaussian_noise),
    ("Erlang", erlang_noise),
    ("Exponential", exp_noise),
    ("Rayleigh", rayleigh_noise)
]

for name, noise in noise_data:
    plt.figure()
    plt.hist(noise.flatten(), bins=50)
    plt.title(name + " Noise Histogram")
    plt.show()

cv2_imshow((uniform_noise * 255).astype(np.uint8))
cv2_imshow((gaussian_noise * 255).astype(np.uint8))
cv2_imshow((gaussian_builtin * 255).astype(np.uint8))
cv2_imshow((erlang_noise * 255).astype(np.uint8))
cv2_imshow((exp_noise * 255).astype(np.uint8))
cv2_imshow((rayleigh_noise * 255).astype(np.uint8))

```
