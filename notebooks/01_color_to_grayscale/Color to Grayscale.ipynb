```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')

path = '/content/drive/MyDrive/Image Processing Lab Files/MST Lab File (.ipynb)/Cat.png'
color = cv2.imread(path, cv2.IMREAD_COLOR)
```

```python
B, G, R = cv2.split(color)

gray = ((B + G + R) / 3).astype(np.uint8)

Bw = float(input("Enter blue: "))
Gw = float(input("Enter green: "))
Rw = float(input("Enter red: "))

gray_weighted = ((Bw*B + Gw*G + Rw*R) / (Bw + Gw + Rw)).astype(np.uint8)

cv2_imshow(gray)
cv2_imshow(gray_weighted)
```
